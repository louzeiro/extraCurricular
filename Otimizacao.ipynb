Chamada da função e criação do solver

```python
import ortools.linear_solver.pywraplp as otlp
```

```python
solver = otlp.Solver('teste',otlp.Solver.GLOP_LINEAR_PROGRAMMING)#otlp.Solver.GLOP_LINEAR_PROGRAMMING)
```

Espaço paramétrico

```python
x = solver.NumVar(0,  # limite inferior
                  10, # limite superior
                  'x' # Nome da variável
                  )

y = solver.NumVar(0, # limite inferior
                  10,# limite superior
                  'y'   
)
```

Restrições

```python
solver.Add(-x+2*y<=8) #restrição 1
solver.Add(2*x+y<=14) #restrição 2
solver.Add(2*x-y<=10) #restrição 3
```

Função Objetivo

```python
solver.Maximize(x+y)
```

Chamada do solver

```python
results = solver.Solve()
```

Exibição do resultado

```python
if results==otlp.Solver.OPTIMAL:
  print(f'A solução ótima foi encontrada para x={x.solution_value()} e y={y.solution_value()}')
```
